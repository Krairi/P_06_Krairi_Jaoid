### Jeu de données entreprise fourni

```python
df_affaire = df_affaire.drop(columns=['name', 'address', 'city', 'state','postal_code', 'latitude', 'longitude', 'stars',
                                     'review_count', 'is_open', 'attributes', 'hours'])
```

### Jeu de données avis des clients fourni

```python
def load_json2(jpath):
    data_iterator = pd.read_json(jpath, chunksize=100000, lines=True)
    chunk_list = []
    for data_chunk in data_iterator:
        filtered_chunk = data_chunk.filter(items=['review_id', 'business_id', 'text', 'stars'])
        chunk_list.append(filtered_chunk)
        print('chunk appended')
        print()

    df = pd.concat(chunk_list)
    return df
```

```python
df_1 = load_json2('yelp_academic_dataset_review.json')
print('Nombre de lignes et de colonnes: ', df_1.shape)
df_1.head()
```

## Créer un seul jeu de données d’avis d’insatisfaction client

Pour créer un seul jeu de données d’avis d’insatisfaction client, j’ai à ma disposition 2 jeux de données créé précédemment, plus le jeu de données importer d’API yelp, j’ai tout d’abord assembler mes 2 premiers jeux de données en utilisant la mèthode merge par l’intermédiaire de 'business_id’, et pour les 2 jeux restant par la mèthode concat, ensuite pour extraire les avis d’insatisfaction client, j’ai utiliser la variable ‘stars’ correspondant au note des clients en lui disant que toute les valeurs inférieurs à 3 correspond à des avis d’insatisfactions client sachant que la valeurs fourni par les clients varies de 0 à 5.

### Assenbler les 2 jeux de données mise à disposition

```python
df_ext= pd.merge(df_affaire, df_1, how='inner', on=['business_id'])
df_ext
```

### Jeu de données avis des clients importer sur l'API Yelp

```python
df_2 = pd.read_csv(data_dir + 'P6_01_fichiercsv_01.csv')
df_2.head()
```

### Assembler le nouveau jeu de données

```python
data=pd.concat([df_ext,df_2], axis=0)
```

```python
data
```

### Sélection uniquement des avis d'insatisfactions

```python
data = data[data.stars < 3]
print('Nombre de lignes et de colonnes: ', data.shape)
data.head()
```

### Supprimer 3 variables

```python
data = data.drop(columns=['business_id', 'review_id','categories'])
data
```
